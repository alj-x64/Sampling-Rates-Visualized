## Visualizing Aliasing and Different Image Sample Rates

```python
import matplotlib.pyplot as plt
from skimage import data, img_as_float, io
from skimage.transform import resize
```

Import the image and define the sample rates you want to use

```python
image = img_as_float(io.imread("mambo.jpg")) #insert your own image

sample_rates = [1, 0.5, 0.25, 0.1]
```

Process the image using skimage and plot it

```python
plt.figure(figsize=(12, 8))
for i, rate in enumerate(sample_rates, 1):

    small = resize(image, (int(image.shape[0]*rate), int(image.shape[1]*rate)),
                   anti_aliasing=False, preserve_range=False)

    plt.subplot(2, 2, i)
    plt.imshow(small)
    plt.title(f"Sample Rate: {int(rate*100)}%")
    plt.axis('off')
plt.suptitle("Aliasing and Sampling Visualization")
plt.show()
```

What if we used anti-aliasing here?

```python
plt.figure(figsize=(12, 8))
for i, rate in enumerate(sample_rates, 1):

    small = resize(image, (int(image.shape[0]*rate), int(image.shape[1]*rate)),
                   anti_aliasing=True, preserve_range=False)

    plt.subplot(2, 2, i)
    plt.imshow(small)
    plt.title(f"Sample Rate: {int(rate*100)}%")
    plt.axis('off')
plt.suptitle("Anti-aliasing and Sampling Visualization")
plt.show()
```
